### Introduction
The WeRateDogs Twitter page has lots of great posts showing peoples' dogs at their best. Luckily, these wonderful pups all have self-supplied ratings so we can understand just how fantastic they are.

Considering some posts have higher self-supplied ratings and others have more likes or retweets, I wondered if there were some factors that may influence that activity. Specifically, I wondered if the self-supplied ratings my be related to retweets and likes, or if the particular dog breed may warrant higher ratings, retweets or likes.

The following insights and graphics demonstrate my finding in this project.

### Insights
1. There seems to be a sweet spot for a rating of 13 in terms of retweet and favorite count.

2. The top 10 highest average rated dog types all have average ratings between 12 - 14, with little variance.

3. In terms of dog types, English Springers, Lakeland Terriers and Standard Poodles all have the most retweets and likes with a significant edge over the other dog types.

```python
import pandas as pd
df_archive_master = pd.read_csv('twitter_archive_master.csv')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df_archive_master['rating_numerator'],df_archive_master['retweet_count'])
plt.xlabel("rating_numerator")
plt.ylabel("retweet_count")
plt.show()
```

```python
plt.scatter(df_archive_master['rating_numerator'],df_archive_master['favorite_count'])
plt.xlabel("rating_numerator")
plt.ylabel("favorite_count")
plt.show()
```

```python
df_archive_master[df_archive_master['p1_dog'] == True].groupby(['p1'])[['rating_numerator','retweet_count','favorite_count']].mean().sort_values(by=['rating_numerator'],ascending=False).head(10).reset_index().plot.bar(x='p1',y='rating_numerator')
```

```python
df_archive_master[df_archive_master['p1_dog'] == True].groupby(['p1'])[['rating_numerator','retweet_count','favorite_count']].mean().sort_values(by=['retweet_count'],ascending=False).head(10).reset_index().plot.bar(x='p1',y='retweet_count')
```

```python
df_archive_master[df_archive_master['p1_dog'] == True].groupby(['p1'])[['rating_numerator','retweet_count','favorite_count']].mean().sort_values(by=['favorite_count'],ascending=False).head(10).reset_index().plot.bar(x='p1',y='favorite_count')
```

### Conclusion
At the end of the day, it appears that certain ratings and dog breeds are associated with greater retweet and like activity. Specifically, it seems that self-supplied ratings of 13/10 tend to bring in the most likes and retweets.

On the other hand, there seems to be less of a significance between self-supplied ratings and particular dog breeds. The top 10 highest rated dog breeds all hover between 12-14 out of 10. These particular breeds aren't necessarily the same top 10 breeds in terms of likes and retweets.

On Average, English Springers, Lakeland Terriers and Standard Poodles all bring in the most retweets and likes.
